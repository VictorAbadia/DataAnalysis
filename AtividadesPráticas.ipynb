```python
# Atividade 1: Sejam os dados da quantidade de visitantes de website lançado recentemente.
# Deseja-se estimar um gráfico de dispersão,considerando os 20 primeiros dias desde o lançamento do website (1-20) e que
# o primeiro dia ele recebeu 20 visitantes e cada dia +20 visitantes do que o dia
# anterior (Dia 1: 20, Dia 2: 40, Dia 3: 60, ...). Use a Regressão Linear Simples.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Criando vetores
dias = range(1, 21) # Variável independente
visitantes = [visitante for visitante in range(20, 420, 20)] # Variável dependente

# Calcula a regressão linear simples
coeficiente_angular, intercepto = np.polyfit(dias, visitantes, 1)

# Coeficientes da regressão
print(f'Coeficiente angular: {coeficiente_angular}')
print(f'Intercepto: {intercepto}')

# Cria previsões usando a equação da regressão
y_pred = coeficiente_angular * dias + intercepto
print(y_pred)

# Criando gráfico de dispersão
plt.scatter(dias, visitantes, label = 'Dados originais')
plt.plot(dias, y_pred, color = 'red', linewidth = 2, label = 'Linha de Regressão')
plt.xlabel('Dias desde o Lançamento')
plt.ylabel('Número de Visitantes')
plt.title('Visitantes por Dia')
plt.grid(True)
plt.show()
```

```python
#Extra - teste
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Criando vetores
dias = range(1, 21)
visitantes = [visitante for visitante in range(20, 420, 20)]

# Criando gráfico de dispersão
plt.scatter(dias, visitantes)
plt.xlabel('Dias desde o Lançamento')
plt.ylabel('Número de Visitantes')
plt.title('Visitantes por Dia')
plt.grid(True)
plt.show()

from scipy.stats import linregress

# Ajustando o modelo de regressão linear
modelo = linregress(dias, visitantes)

# Obtendo coeficientes
inclinação = modelo.slope
interceptação = modelo.intercept

# Criando vetor de dias para previsão
dias_previsao = range(21, 26)

# Calculando valores previstos
visitantes_previstos = [inclinação * dia + interceptação for dia in dias_previsao]

# Plotando reta de regressão
plt.plot(dias, inclinação * dias + interceptação, color='red')

# Adicionando marcadores para valores previstos
plt.scatter(dias_previsao, visitantes_previstos, marker='o', color='green', label='Valores Previstos')

# Ajustando legenda
plt.legend()

# Mostrando o gráfico
plt.show()
```

```python
# Atividade 2: Análise de Valor de casas com base no número de garagens que estas possuem.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

dados = pd.read_csv('HousePrices_HalfMil.csv', sep=',')
analise = dados[['Garage','Prices']]

preco = dados[['Prices']] # Variável Dependente
garagem = dados[['Garage']] # Variável Independente

# Corrigindo os arrays para 1D (erro de sintaxe)
preco = analise['Prices'].to_numpy().ravel()
garagem = analise['Garage'].to_numpy().ravel()

# Calcula a regressão linear simples
coeficiente_angular, intercepto = np.polyfit(preco, garagem, 1)

# Coeficientes da regressão
print(f'Coeficiente angular: {coeficiente_angular}')
print(f'Intercepto: {intercepto}')

# Cria previsões usando a equação da regressão
y_pred = coeficiente_angular * preco + intercepto
print(y_pred)

# Criando gráfico de dispersão
plt.scatter(preco, garagem, label = 'Dados originais')
plt.plot(preco, y_pred, color = 'red', linewidth = 2, label = 'Linha de Regressão')
plt.xlabel('Valor do Imóvel')
plt.ylabel('Número de Garagens')
plt.title('Preço X Garagem')
plt.grid(True)
plt.show()
```
